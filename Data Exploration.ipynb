```python
PATH = 'data/'

def read_data(reader,PATH):
#     reader.bundle_axes = 'yxz'
#     reader.default_coords = 1
    data = []
    for each in listdir(PATH):
        data.append(reader(PATH+each))
    return data
```

```python
data = read_data(nd2reader.ND2Reader,PATH)
len(data)
```

```python
sample = data[0]
sample.metadata
```

```python
sample.sizes
```

```python
sample.bundle_axes = 'yx'
sample.default_coords['c'] = 1
sample.default_coords['z'] = 10
```
