```python
import nd2reader
import matplotlib.pyplot as plt
import os
from os import listdir
from PIL import Image
```

```python
PATH = 'data/'

def read_data(reader,PATH):
    data = []
    for each in listdir(PATH):
        data.append(reader(PATH+each))
    return data
```

```python
data = read_data(nd2reader.Nd2,PATH)
len(data)
```

```python
sample = data[0]
sample
```

```python
for i in sample:
    plt.imshow(i,cmap=plt.cm.gray)
    plt.show()
    plt.close()
```
